```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

# LS_12(1)

```python
df = pd.read_csv('LS_12(1).csv')
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, shuffle=False)
```

### Polynomial regression

```python
poly_model = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_model.fit_transform(X_train)
regr = LinearRegression()
regr.fit(X_poly, y_train)
```

```python
train_pred = regr.predict(X_poly)
test_pred = regr.predict(poly_model.fit_transform(X_test))
```

```python
plt.figure()
plt.plot(df['cycle'], df['capacitance'], label='Actual', color='royalblue')
plt.plot(df['cycle'].head(len(train_pred)), train_pred, label='Training', color='g')
plt.plot(df['cycle'].tail(len(test_pred)), test_pred, label='Forecast', color='red')
plt.xlabel('Cycle No.')
plt.ylabel(r'Specific Capacitance (uF/cm$^2$)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print("training r^2 score: ", r2_score(y_train, train_pred))
print("forecast r^2 score: ", r2_score(y_test, test_pred))
```

```python
plt.figure()
plt.plot(df['capacitance'], df['capacitance'], color='k')
plt.scatter(df['capacitance'].head(len(train_pred)), train_pred, color='g', label='training')
plt.scatter(df['capacitance'].tail(len(test_pred)), test_pred, color='red', label='forecast')
plt.xlabel(r'Actual Specific Capacitance (uF/cm$^2$)')
plt.ylabel(r'Predicted Specific Capacitance (uF/cm$^2$)')
plt.legend()
plt.tight_layout()
plt.show()
```

### Logarithmic regression

```python
fit = np.polyfit(np.log(X_train.iloc[:,0]), y_train, 1)
y1 = fit[0]*np.log(np.arange(1,len(y_train)+1)) + fit[1]
fit = np.polyfit(np.log(X_test.iloc[:,0]), y_test, 1)
y2 = fit[0]*np.log(np.arange(len(y_train)+1,len(df)+1)) + fit[1]
```

```python
plt.figure()
plt.plot(df['cycle'], df['capacitance'], label='Actual', color='royalblue')
plt.plot(df['cycle'].head(len(y1)), y1, label='Training', color='g')
plt.plot(df['cycle'].tail(len(y2)), y2, label='Forecast', color='red')
plt.xlabel('Cycle No.')
plt.ylabel(r'Specific Capacitance (uF/cm$^2$)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print("training r^2 score: ", r2_score(y_train, y1))
print("forecast r^2 score: ", r2_score(y_test, y2))
```

```python
plt.figure()
plt.plot(df['capacitance'], df['capacitance'], color='k')
plt.scatter(df['capacitance'].head(len(y1)), y1, color='g', label='training')
plt.scatter(df['capacitance'].tail(len(y2)), y2, color='red', label='forecast')
plt.xlabel(r'Actual Specific Capacitance (uF/cm$^2$)')
plt.ylabel(r'Predicted Specific Capacitance (uF/cm$^2$)')
plt.legend()
plt.tight_layout()
plt.show()
```

### Exponential Regression

```python
fit = np.polyfit(X_train.iloc[:,0], np.log(y_train), 1)
y1 = np.exp(fit[1]) * np.exp(fit[0]*np.arange(1,len(y_train)+1))
fit = np.polyfit(X_test.iloc[:,0], np.log(y_test), 1)
y2 = np.exp(fit[1]) * np.exp(fit[0]*np.arange(len(y_train)+1,len(df)+1))
```

```python
plt.figure()
plt.plot(df['cycle'], df['capacitance'], label='Actual', color='royalblue')
plt.plot(df['cycle'].head(len(y1)), y1, label='Training', color='g')
plt.plot(df['cycle'].tail(len(y2)), y2, label='Forecast', color='red')
plt.xlabel('Cycle No.')
plt.ylabel(r'Specific Capacitance (uF/cm$^2$)')
plt.legend()
plt.tight_layout()
plt.show()
```
