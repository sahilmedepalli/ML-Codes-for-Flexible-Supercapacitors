```python
import pandas as pd
import numpy as np
import sklearn.gaussian_process as gp
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

# LS_12(1)

```python
df = pd.read_csv('LS_12(1).csv')
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, shuffle=False)
```

```python
kernel = gp.kernels.ConstantKernel(1.0, (0.1, 1000)) * gp.kernels.RBF(10.0, (0.001, 1000))
model = gp.GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, alpha=0.1, normalize_y=True)
```

```python
model.fit(X_train, y_train)
params = model.kernel_.get_params()
```

```python
train_pred, stdtr = model.predict(X_train, return_std=True)
test_pred, stdte = model.predict(X_test, return_std=True)
citest = 1.96 * stdte
```

```python
fig, ax = plt.subplots()
ax.plot(df['cycle'], df['capacitance'], 'b--', label='Actual')
ax.plot(X_train['cycle'], train_pred, 'g', label='Training')
ax.plot(X_test['cycle'], test_pred, 'r:', label='Forecast')
ax.fill_between(X_test['cycle'], (test_pred-citest), (test_pred+citest), color='r', alpha=0.1)
plt.xlabel('Cycle No.')
plt.ylabel(r'Specific Capacitance (uF/cm$^2$)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print("training r^2 score: ", r2_score(y_train, train_pred))
print("forecast r^2 score: ", r2_score(y_test, test_pred))
```
